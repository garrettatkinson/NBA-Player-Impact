## Predicting Player VORP Using Main Box Score Statistics

```python
# Importing Linear Regression Model Package
from sklearn.linear_model import LinearRegression
```

```python
# Response Variable
Y = nba_2022_data['VORP']
```

```python
# Total Boxscore Stats
X = nba_2022_data[['PTS', 'AST', 'TRB', 'STL', 'BLK']]
```

```python
reg1 = LinearRegression().fit(X, Y)
reg1.score(X, Y)
```

```python
# Per Game Stats
X2 = nba_2022_data[['PTS Per Game', 'AST Per Game', 'TRB Per Game', 'STL Per Game', 'BLK Per Game']]
```

```python
reg2 = LinearRegression().fit(X2, Y)
reg2.score(X2, Y)
```

```python
# Per 36 Minutes Stats
X3 = nba_2022_data[['PTS Per 36', 'AST Per 36', 'TRB Per 36', 'STL Per 36', 'BLK Per 36']]
```

```python
reg3 = LinearRegression().fit(X3, Y)
reg3.score(X3, Y)
```

```python
# Per 100 Possessions Stats
X4 = nba_2022_data[['PTS Per 100', 'AST Per 100', 'TRB Per 100', 'STL Per 100', 'BLK Per 100']]
```

```python
reg4 = LinearRegression().fit(X4, Y)
reg4.score(X4, Y)
```

```python
# Total Stats Had the Best, therefore we will use reg1
Y_hat = reg1.predict(X)
```

```python
import plotly.express as px
# new
import plotly.graph_objects as go
```

```python
# Predicted vs Actual VORP Using Total Stats

fig = px.scatter(
    x=Y,
    y=Y_hat,
    color=nba_2022_data.index,
    size=nba_2022_data['MP Per Game'],
    labels = {
        "x": "Player VORP",
        "y": "Predicted Player VORP"
    }
)

fig.add_trace(
    go.Scatter(
        x=Y,
        y=Y,
        mode="lines"
    )
)

fig.show()
```

## Are Players Being Paid Correctly?

```python
# Bring in NBA Contract Data for 2023-2024 Season
nba_contracts = pd.read_csv('./data/nba_player_contracts.csv', index_col= 'Player').fillna(0)
nba_contracts = nba_contracts.drop(columns=['2024-25', '2025-26', '2026-27', '2027-28', '2028-29'])
nba_contracts.head()
```

```python
merged_nba = pd.merge(nba_2022_data, nba_contracts, on="Player")
merged_nba = merged_nba.drop(columns = ['-9999', 'Rk_y', 'Tm_y'])
merged_nba.head()
```

```python
# Seeing Who is the Highest Paid Player in 2023-2024
merged_nba["2023-24"].idxmax()
```

```python
# Filter out players with negative VORP
merged_nba_pos = merged_nba[merged_nba['VORP'] > 0]
Xpos = merged_nba_pos.drop(columns = ['2023-24', 'Guaranteed', 'Rk_x', 'Pos', 'Tm_x', 'PlayerID', 'PG%', 'SG%', 'SF%', 'PF%', 'C%'], axis = 1)
Ypos = merged_nba_pos['Guaranteed']
```

```python
money_reg = LinearRegression().fit(Xpos, Ypos)
money_reg.score(Xpos, Ypos)
```

```python
Y_hat_pos = money_reg.predict(Xpos)
```

```python
# Predicted vs Actual Guaranteed Money

fig = px.scatter(
    x=Ypos,
    y=Y_hat_pos,
    color=merged_nba_pos.index,
    size=merged_nba_pos['VORP'],
    labels = {
        "x": "Total Guaranteed Money",
        "y": "Predicted Guaranteed Money"
    }
)

fig.add_trace(
    go.Scatter(
        x=Ypos,
        y=Ypos,
        mode="lines"
    )
)

fig.show()
```

```python
# With Positive VORP
Xpos2 = merged_nba_pos.drop(columns = ['2023-24', 'Guaranteed', 'Rk_x', 'Pos', 'Tm_x', 'PlayerID', 'PG%', 'SG%', 'SF%', 'PF%', 'C%'], axis = 1)
Ypos2 = merged_nba_pos['2023-24']
money_reg_pos2 = LinearRegression().fit(Xpos2, Ypos2)
money_reg_pos2.score(Xpos2, Ypos2)
```

```python
Y_hat_pos2 = money_reg_pos2.predict(Xpos2)
```

```python
# 2023-24 Money vs Predicted 2023-24 Money
fig = px.scatter(
    x=Ypos2,
    y=Y_hat_pos2,
    color=merged_nba_pos.index,
    size=merged_nba_pos['VORP'],
    labels = {
        "x": "2023-24 Money",
        "y": "Predicted 2023-24 Money"
    }
)

fig.add_trace(
    go.Scatter(
        x=Ypos2,
        y=Ypos2,
        mode="lines"
    )
)
```
